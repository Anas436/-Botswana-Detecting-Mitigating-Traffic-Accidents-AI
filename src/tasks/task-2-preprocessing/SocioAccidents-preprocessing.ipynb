```python
import pandas as pd 
```

```python
df = pd.read_csv('../../data/preprocessed_data/botswana-gdp.csv')
```

```python
df
```

```python
df.info()
```

```python
df_inflation = pd.read_csv('../../data/preprocessed_data/botswana-inflation-rate.csv')
```

```python
df_inflation.head()
```

```python
merged_df = df_inflation.merge(df, on='date')
```

```python
#lower all columns and remove white spaces
merged_df.columns = merged_df.columns.str.lower().str.replace(' ', '_')
```

```python
merged_df.head()
```

```python
#remove _ at the beginning of each column name
for column in df.columns:
    if column.startswith('_'):
        new_column = column.lstrip('_')  # Remove the underscore at the beginning
        df.rename(columns={column: new_column}, inplace=True)
```

```python
merged_df.head()
```

```python
to_remove = ['_annual_change', '_annual_%_change']
merged_df = merged_df.drop(columns=to_remove)
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
merged_df['date'] = pd.to_datetime(merged_df['date'])
```

```python
merged_df['year'] = merged_df['date'].dt.year
```

```python
merged_df.head()
```

```python
merged_df = merged_df.drop(columns=['date'])
```

```python
merged_df.head()
```

```python
merged_df = merged_df[['year'] + [col for col in merged_df.columns if col != 'year']]
```

```python
merged_df.head()
```

```python
crash_trends = pd.read_csv("../../data/preprocessed_data/crash_trends_1981-2021.csv")
```

```python
crash_trends.head()
```

```python
crash_trends = crash_trends.drop(columns=['Unnamed: 0'])
```

```python
crash_trends.head()
```

```python
crash_trends['year'] = pd.to_datetime(crash_trends['year']).dt.year
```

```python
crash_trends.head()
```

```python
accident_df = crash_trends.merge(merged_df, on='year')
```

```python
accident_df.head()
```

```python
accident_df.T
```

```python
accident_df.info()
```

```python
new_column = {
    '_inflation_rate_(%)': 'inflation_rate_(%)',
    '_gdp_(_billions_of_us_$)': 'gdp_(_billions_of_us_$)',
    '_per_capita_(us_$)': 'per_capita_($)'
}
accident_df.rename(columns=new_column, inplace=True)
```

```python
accident_df.info()
```

```python
accident_df.to_csv('../../data/preprocessed_data/botswanaSocioAccidents.csv')
```
